```python
import pandas as pd
import numpy as np
from faker import Faker  

fake = Faker(['es', 'es_ES', 'es_MX'])

def cliente(num):

    estado_civil = ['Soltero', 'Casado', 'Union libre', 'Separado', 'Viudo', 'No revelado']
    dispositivo_movil = ['Android', 'iOS', 'Otro', 'No tiene']
    cel_prefix = [300, 301, 302, 304, 305, 303, 304, 305, 310, 311, 312, 313, 314, 320, 321, 322, 323, 315, 316, 317, 318, 319, 350, 351]

    team_list = [fake.color_name() for x in range(4)]

    fake_cliente = [{
                        'codigo_cliente':x + 100000,
                        'nombre':fake.first_name(),
                        'apellido':fake.last_name() + ' ' + fake.last_name(),
                        'cedula':1234,
                        'estado_civil':np.random.choice(estado_civil, p=[0.40, 0.30, 0.15, 0.05, 0.05, 0.05]),
                        'email':fake.ascii_email(),
                        'celular':'+57' + str(np.random.choice(cel_prefix)) + str(np.random.randint(3000000, 9000000 + 1)),
                        'edad':00,
                        'dispositivo_movil':np.random.choice(dispositivo_movil, p=[0.55, 0.35, 0.05, 0.05]),
                        'perfil_digital':00,
                        'clave_dinamica':fake.boolean(chance_of_getting_true=65),
                        'cant_prod_activos':0,
                        'acceso_internet':fake.boolean(chance_of_getting_true=75),
                        'saldo':0,
                        'dias_cartera':0,
                        'saldo_cartera':0, # Si dias_cartera > 0
                        'tam_comercial':'pequeño'
                    } for x in range(num)]

    return fake_cliente

worker_df = pd.DataFrame(cliente(5000000))
worker_df.head()
```

```python
from faker import Faker     
import numpy as np
import pandas as pd

industry = ['Automotive','Health Care','Manufacturing','High Tech','Retail']

# Instantiate Faker with multiple locales
fake = Faker(['en_US','de_DE','pt_BR','ja_JP','zh-CN']) 
def create_data(x): 
  
    # dictionary 
    b_user ={} 
    for i in range(0, x): 
        b_user[i] = {} 
        b_user[i]['name'] = fake.name()
        b_user[i]['job'] = fake.job()
        b_user[i]['birthdate'] = fake.date_of_birth(minimum_age=18,maximum_age=65)
        b_user[i]['email'] = fake.company_email()
        b_user[i]['company'] = fake.company()
        b_user[i]['industry'] = fake.random_element(industry)
        b_user[i]['city'] = fake.city()
        b_user[i]['state'] = fake.state()
        b_user[i]['zipcode'] = fake.postcode()
        b_user[i]['netNew'] = fake.boolean(chance_of_getting_true=65)
        b_user[i]['sales_rounded'] = round(np.random.normal(1000,200))
        b_user[i]['sales_decimal'] = np.random.normal(1000,200)
        b_user[i]['priority'] = fake.random_digit()
        b_user[i]['industry2'] = np.random.choice(industry)        
        
    return b_user
    
df = pd.DataFrame(create_data(1000)).transpose()
# df = pd.DataFrame(create_data(1000))
df.head(10)
```
